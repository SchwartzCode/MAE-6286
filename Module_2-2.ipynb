```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
def linear_convection(nx, L=2.0, c=1.0, dt = 0.025, nt=20):
    dx = L / (nx - 1)
    x = np.linspace(0.0, L, num=nx)
    u0 = np.ones(nx)
    mask = np.where(np.logical_and(x>=0.5, x <= 1.0))
    u0[mask] = 2.0
    u = u0.copy()
    
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt / dx * (u[1:] - u[:-1])
        
    plt.figure(figsize=(4.0,4.0))
    plt.xlabel('x')
    plt.ylabel('u')
    plt.grid()
    plt.plot(x, u0, label='Initial', color='purple', linestyle='--', linewidth=2)
    plt.plot(x, u, label='nt = {}'.format(nt), color='r', linestyle='-', linewidth=2)
    plt.legend()
    plt.xlim(0.0, L)
    plt.ylim(0.0, 2.5)
```

```python
linear_convection(41)
linear_convection(41, nt=12)
```

```python
def linear_convection_cfl(nx, L=2.0, c=1.0, sigma=0.5, nt=20):
    dx = L / (nx - 1)
    x = np.linspace(0.0, L, num=nx)
    dt = sigma * dx / c
    u0 = np.ones(nx)
    mask = np.where(np.logical_and(x>=0.5, x <= 1.0))
    u0[mask] = 2.0
    u = u0.copy()
    
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt / dx * (u[1:] - u[:-1])
        
    plt.figure(figsize=(4.0,4.0))
    plt.xlabel('x')
    plt.ylabel('u')
    plt.grid()
    plt.plot(x, u0, label='Initial', color='purple', linestyle='--', linewidth=2)
    plt.plot(x, u, label='nt = {}'.format(nt), color='r', linestyle='-', linewidth=2)
    plt.legend()
    plt.xlim(0.0, L)
    plt.ylim(0.0, 2.5)
```

```python
linear_convection_cfl(180)
linear_convection_cfl(180, sigma=1.1)
```
